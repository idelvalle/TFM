```python
%%R
library(datamicroarray)
data('pomeroy', package = 'datamicroarray')
data('singh', package='datamicroarray')
```

```python
%%R
str(pomeroy)
str(singh)
```

```python
%%R

table(pomeroy$y)
```

```python
%%R

table(singh$y)
```

```python
%%R

cns = as.data.frame(pomeroy)
cns$y = unclass(cns$y)

prostate = as.data.frame(singh)
prostate$y = unclass(prostate$y)

```

```python
%%R

write.table(cns, file = "cns.csv", sep=",")
write.table(prostate, file = "prostate.csv", sep=",")
```
